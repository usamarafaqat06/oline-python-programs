```python
eng = int(input("Enter marks of English"));
isl = int(input("Enter marks of Islamiyat"));
math = int(input("Enter marks of Math"));
urdu = int(input("Enter marks of Urdu"));
phy = int(input("Enter marks of Physics"));
totalMarks = 500;
print("Total Marks = ", totalMarks);
obtainMarks = eng + isl + math + urdu + phy;
print("Obtain Marks = ", obtainMarks);
percentage = (obtainMarks/totalMarks) * 100;
print("Percentage = ", percentage);
if (eng > 100 or eng < 0):
    print("invalid english marks! please enter valid marks lies between 0 to 100 ");
elif (isl > 100 or isl < 0):
    print("invalid islamiyat marks! please enter valid marks lies between 0 to 100 ");
elif (math > 100 or math < 0):
    print("invalid math marks! please enter valid marks lies between 0 to 100 ");
elif (urdu > 100 or urdu < 0):
    print("invalid urdu marks! please enter valid marks lies between 0 to 100 ");
elif (phy > 100 or phy < 0):
    print("invalid physics marks! please enter valid marks lies between 0 to 100 ");
elif (percentage <= 100 and percentage >= 90):
    print("Grade A+");
elif (percentage < 90 and percentage >= 80):
    print("Grade A"); 
elif (percentage < 80 and percentage >= 70):
    print("Grade B+");
elif (percentage < 70 and percentage >= 60):
    print("Grade B");
elif (percentage < 60 and percentage >= 50):
    print("Grade C");
elif (percentage < 50 and percentage >= 40):
    print("Grade D");
elif (percentage < 40 and percentage >= 33):
    print("Grade E");
else:
    print("Fail");
       
```
